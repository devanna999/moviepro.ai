Let's keep everything in a function

```python
all_stopwords = ['i', 'me', 'my', 'myself', 'we', 'our', 'ours','as','ourselves', 'you', "you're", "you've", "you'll", "you'd", 'your', 'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', 'she', "she's", 'her', 'hers', 'herself', 'it', "it's", 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves', 'what', 'which', 'who', 'whom', 'this', 'that', "that'll", 'these', 'those', 'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'having', 'do', 'does', 'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until', 'while', 'of', 'at', 'by', 'for', 'with', 'about', 'against', 'between', 'into', 'through', 'during', 'before', 'after', 'above', 'below', 'to', 'from', 'up', 'down', 'in', 'out', 'on', 'off', 'over', 'under', 'again', 'further', 'then', 'once', 'here', 'there', 'when', 'where', 'why', 'how', 'all', 'any', 'both', 'each', 'few', 'more', 'most', 'other', 'some', 'such', 'no', 'nor', 'not', 'only', 'own', 'same', 'so', 'than', 'too', 'very', 's', 't', 'can', 'will', 'just', 'don', "don't", 'should', "should've", 'now', 'd', 'll', 'm', 'o', 're', 've', 'y', 'ain', 'aren', "aren't", 'couldn', "couldn't", 'didn', "didn't", 'doesn', "doesn't", 'hadn', "hadn't", 'hasn', "hasn't", 'haven', "haven't", 'isn', "isn't", 'ma', 'mightn', "mightn't", 'mustn', "mustn't", 'needn', "needn't", 'shan', "shan't", 'shouldn', "shouldn't", 'wasn', "wasn't", 'weren', "weren't", 'won', "won't", 'wouldn', "wouldn't"]

def get_term_frequency(input_text):
  wordlist = re.findall(r"[\w']+", input_text)
  clean_wordlist = [w.lower() for w in wordlist if w.lower() not in all_stopwords]
  cleanwordcounts = Counter(clean_wordlist)
  cleanwordcounts = dict(cleanwordcounts)
  # total_count = sum(cleanwordcounts.values(), 0.0)
  total_count = len(wordlist)
  normalized_cleanwordcounts = {k: float("{:.5f}".format(v / total_count))  for k, v in cleanwordcounts.items()}
  return normalized_cleanwordcounts
```

```python
normalized_cleanwordcounts_batmanbegins_plt = get_term_frequency(batmanbegins_plot)
print (normalized_cleanwordcounts_batmanbegins_plt)
```

Term frequency indicates how popular a word is within a given document. But even a stopword that you failed to remove or a generic word in a different domain might have high term frequency in a given document.

Let's understand the characteristics of an important words.

A word is important or called a "keyword" if it occurs very frequently in a document but it rarely appears in other document. 

So keywords have high frequency within a document (Term Frequency) and low frequency across documents (document frequency)

So importance of a word is directly proportinal to TF and inversely proportional to document frequency.

So importance of a word =  TF * (1/DF)

Lets call 1/DF as IDF (Inverse document Frequency)


So importance of a word is TF*IDF

IDF_of_a_word = 1/document_frequency_of_a_word = N/ document_counter


```python
IDF_of_a_word = N/document_counter  # Varies between 1 to N.
print (IDF_of_a_word)
```

```python
# Let's avoid divide by zero errors. We add extra one at top and bottom to avoid divide by zero errors etc.

IDF_of_a_word = ((1+N)/(document_counter+1))
print (IDF_of_a_word)
```

```python
# We add an extra one to total sum after log because if a word is present in all we get IDF = 0 otherwise.
import math
IDF = math.log10( 1 )
print(IDF)

IDF = math.log10( 10 )
print(IDF)

IDF = math.log10( 1000 )
print(IDF)

IDF = math.log10( 10000 )
print(IDF)

IDF = math.log10( IDF_of_a_word )+1
print(IDF)
```

```python

def get_inverse_document_frequency(word, full_dataframe):
  wrd = word.lower()
  document_counter = 0
  for index, row in full_dataframe.iterrows():
    if testword in list(row['TF'].keys()):
      document_counter = document_counter +1

  total_number_of_documents = full_dataframe.shape[0]
  IDF_of_word = (1+total_number_of_documents)/(document_counter+1)
  IDF = math.log10( IDF_of_word )+1
  return IDF
```

Basically for each word in a document we calculated an importance factor called TFIDF just based on statistics of that word in that document and in the whole corpus.

## TF-IDF using Sklearn

https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
# fit calculates all unique words after stop word removal and also calculates each word's IDF values.
tfidfvectorizer = TfidfVectorizer(analyzer='word',stop_words= 'english')
tfidfvectorizer.fit(df['Plot'])
```

```python
ftr_names =  tfidfvectorizer.get_feature_names()
print (ftr_names)
print (len(ftr_names))
```

```python
print(tfidfvectorizer.idf_)
tfidfvectorizer.idf_.shape
```

```python
# transform calculates the TFIDF for each document. It calculates TF for each word, multiplies it with the IDF calculated previously to get TFIDF.
transform_output = tfidfvectorizer.transform(df['Plot'])
```

```python
transform_output.shape
```

```python
tfidf_df = pd.DataFrame(transform_output.toarray(), columns=ftr_names)
tfidf_df.head(30)
```
